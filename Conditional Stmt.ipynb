```python
units = int(input("Enter number of units consumed: "))
bill = 0

if units <= 100:
    bill = 0

if units > 100 and units <= 200:
    bill = (units - 100) * 5

if units > 200:
    bill = (100 * 5) + (units - 200) * 10

print("Electricity Bill = Rs", bill)
```

```python
#Accept the marked price from the user and calculate the Net amount as (Marked Price â€“ Discount) to pay  according to following criteria:

price = int(input("Marked Price enter karo: "))

if price > 10000:
    discount = price * 20 / 100      
elif price > 7000:
    discount = price * 15 / 100      
else:
    discount = price * 10 / 100      

net_price = price - discount

print("Discount:", discount)
print("Net amount to pay:", net_price)

```

```python
#Accept the marks of English, Math and Science,Social Studies Subject and display the stream allotted

eng = int(input("Enter English marks: "))
math = int(input("Enter Math marks: "))
sci = int(input("Enter Science marks: "))
sst = int(input("Enter Social Studies marks: "))

if eng > 80 and math > 80 and sci > 80 and sst > 80:
    print("Stream Allotted: Science Stream")

elif eng > 80 and math > 50 and sci > 50:
    print("Stream Allotted: Commerce Stream")

elif eng > 80 and sst > 80:
    print("Stream Allotted: Humanities Stream")

else:
    print("No stream allotted")
```
